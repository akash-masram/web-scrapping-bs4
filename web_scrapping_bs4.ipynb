```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Load the HTML (replace with your actual URL or local HTML file)
url = "https://www.imageanime.com/"  # Replace with actual page URL
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Step 2: Find all the product cards
cards = soup.find_all("td", class_="horizontal-seperator vertical-seperator")

# Step 3: Extract data
names = []
prices = []
sale_prices = []
img_urls = []

for card in cards:
    name = card.find('div', class_='name').get_text(strip=True)
    price = card.find('div', class_='price').get_text(strip=True).replace("Price: ", "")
    sale_price = card.find('div', class_='sale-price-bold').get_text(strip=True).replace("SALE PRICE: ", "")
    img_url = card.find('img')['src']

    names.append(name)
    prices.append(price)
    sale_prices.append(sale_price)
    img_urls.append(img_url)

# Step 4: Convert to DataFrame
df = pd.DataFrame({
    'Name': names,
    'Price': prices,
    'Sale Price': sale_prices,
    'Image URL': img_urls
})

# Step 5: Save to CSV
df.to_csv("products.csv", index=False)

print("CSV file 'products.csv' has been created successfully!")
```
